```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate 10 million heights: normal dist with mean 170 cm, sd 10 cm
num_people = 10000000
heights = np.random.normal(170, 10, num_people)

# Verify stats
mean_height = np.mean(heights)
sd_height = np.std(heights)

print(f"Simulated Mean: {mean_height:.2f} cm")
print(f"Simulated SD: {sd_height:.2f} cm")

# Generate histogram
plt.figure(figsize=(10, 6))
plt.hist(heights, bins=100, density=True, alpha=0.7, color='blue')
plt.title('Histogram of Simulated Heights for 10 Million People in Telangana')
plt.xlabel('Height (cm)')
plt.ylabel('Density')
plt.axvline(mean_height, color='red', linestyle='--', label=f'Mean = {mean_height:.2f} cm')
plt.axvline(mean_height + sd_height, color='green', linestyle='--', label=f'+1 SD = {mean_height + sd_height:.2f} cm')
plt.axvline(mean_height - sd_height, color='green', linestyle='--', label=f'-1 SD = {mean_height - sd_height:.2f} cm')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Specific sample heights
heights = np.array([165.2, 172.5, 168.1, 175.3, 160.8, 179.9, 169.4, 171.0, 173.7, 167.6])

# Compute mean and sample SD
mean_height = np.mean(heights)
std_dev = np.std(heights, ddof=1)  # ddof=1 for sample SD (divide by n-1)

print(f"Sample Mean: {mean_height:.2f} cm")
print(f"Sample SD: {std_dev:.2f} cm")

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(heights, bins=5, alpha=0.7, color='blue')  # Fewer bins for small n
plt.title('Histogram of Sample Heights (n=10)')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.axvline(mean_height, color='red', linestyle='--', label=f'Sample Mean = {mean_height:.2f} cm')
plt.axvline(170, color='green', linestyle='--', label='Population Mean = 170 cm')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Simulate a large population of heights (10 million people)
np.random.seed(42)  # for reproducibility
population_size = 10000000
population_heights = np.random.normal(loc=170, scale=10, size=population_size)

print("Population mean:", np.mean(population_heights))
print("Population std:", np.std(population_heights))

# Step 2: Draw m=1 sample of size n=25
n = 25
sample = np.random.choice(population_heights, size=n, replace=True)  # One sample

# Compute sample mean and sample standard deviation (ddof=1 for sample sd)
sample_mean = np.mean(sample)
sample_sd = np.std(sample, ddof=1)

# Step 3: Report results
print("\n=== Single Sample (n=25) ===")
print("Sample heights:", sample)
print("Sample mean:", sample_mean)
print("Sample s.d.:", sample_sd)
print("Standard error of the mean (theoretical):", 10 / np.sqrt(n))  #  2.0 for n=25

# Optional: Histogram of the single sample
plt.figure(figsize=(8, 5))
plt.hist(sample, bins=10, alpha=0.7, color='blue')
plt.title('Histogram of One Sample of 25 Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Sample Mean = {sample_mean:.2f} cm')
plt.axvline(170, color='green', linestyle='--', label='Population Mean = 170 cm')
plt.legend()
plt.show()
```
