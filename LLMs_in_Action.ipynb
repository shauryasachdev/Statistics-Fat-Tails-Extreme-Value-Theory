```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
max_n = 500   # Maximum sample size
true_mean = 0.5 # True mean for fair coin

# Simulate single run
plt.figure(figsize=(10, 6))
# Generate coin tosses: 1 (heads) or 0 (tails)
samples = np.random.choice([0, 1], size=max_n)
# Compute cumulative means
cumulative_means = np.cumsum(samples) / np.arange(1, max_n + 1)
plt.plot(cumulative_means, label='Single Run')

# Plot true mean
plt.axhline(true_mean, color='red', linestyle='--', label='True Mean (0.5)')
plt.xlabel('Sample Size n')
plt.ylabel('Sample Mean')
plt.title('WLLN: Coin Toss Sample Mean Converging to 0.5 (1st Run)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_trials = 5          # Number of independent trials to show
max_n = 300             # Maximum sample size (number of dice rolls per trial)
true_mean = 3.5         # True mean for a fair six-sided die

# Create the plot
plt.figure(figsize=(12, 7))

# Run 5 independent trials
for trial in range(num_trials):
    # Generate 300 dice rolls: integers from 1 to 6
    samples = np.random.randint(1, 7, size=max_n)

    # Compute the running (cumulative) sample means
    cumulative_means = np.cumsum(samples) / np.arange(1, max_n + 1)

    # Plot this trial's convergence path
    plt.plot(cumulative_means, alpha=0.7, linewidth=1.5,
             label=f'Trial {trial+1}')

# Add the true mean line
plt.axhline(true_mean, color='red', linestyle='--', linewidth=2.5,
            label=f'True Mean = {true_mean}')

# Formatting
plt.xlabel('Number of Dice Rolls (n)', fontsize=12)
plt.ylabel('Running Sample Mean', fontsize=12)
plt.title('Weak Law of Large Numbers\nDice Rolls (16)  5 Independent Trials up to n=300',
          fontsize=14, fontweight='bold')
plt.legend(loc='upper right', fontsize=10)
plt.grid(True, alpha=0.3)
plt.ylim(2.0, 5.0)  # Optional: zoom in around the interesting range
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lam = 1.0                  # Rate parameter
true_mean = 1 / lam        # True mean = 1.0
num_trials = 5             # Number of independent trials
max_n = 300             # Maximum sample size

# Create single figure
plt.figure(figsize=(12, 7))

# Run 5 independent trials
for trial in range(num_trials):
    # Generate exponential random samples (scale = 1/lambda = mean)
    samples = np.random.exponential(scale=1/lam, size=max_n)

    # Compute running (cumulative) sample means
    cumulative_means = np.cumsum(samples) / np.arange(1, max_n + 1)

    # Plot this trial
    plt.plot(cumulative_means, alpha=0.7, linewidth=1.4,
             label=f'Trial {trial+1}')

# Add true mean reference line
plt.axhline(true_mean, color='red', linestyle='--', linewidth=2.5,
            label=f'True Mean = {true_mean:.1f}')

# Formatting
plt.xlabel('Sample Size n', fontsize=12)
plt.ylabel('Running Sample Mean', fontsize=12)
plt.title(f'Weak Law of Large Numbers  Exponential Distribution\n'
          f'( = {lam},  = {true_mean:.1f})  5 Independent Trials up to n={max_n:,}',
          fontsize=14, fontweight='bold')
plt.legend(loc='upper right', fontsize=10)
plt.grid(True, alpha=0.3)

# Optional: zoom y-axis around the true mean for clearer view
plt.ylim(true_mean * 0.4, true_mean * 1.6)

plt.tight_layout()
plt.show()
```
